```python
query = """
SELECT
  ROW_NUMBER() OVER (ORDER BY prd.start_date, prd.product_name) AS product_key,
  prd.product_id,
  prd.product_number,
  prd.product_name,
  prd.product_line,
  prd.category_id,
  prd_cat.category,
  prd_cat.sub_category,
  prd.start_date,
  CASE 
    WHEN prd.end_date IS NULL THEN true 
    ELSE false 
  END AS active_product,
  prd_cat.maintenance_flag
FROM baraa_projectwork.silver.crm_products AS prd
LEFT JOIN baraa_projectwork.silver.erp_product_category AS prd_cat
  ON prd.category_id = prd_cat.category_id
"""
df = spark.sql(query)
```

```python
df.limit(10).display()
```

```python
df.write.mode("overwrite").format("delta").saveAsTable("baraa_projectwork.gold.dim_products")
```
