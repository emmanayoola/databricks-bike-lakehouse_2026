```python
query = """
SELECT
  sales.order_number,
  products.product_key,
  customers.customer_key,
  sales.order_date,
  sales.ship_date,
  sales.due_date,
  sales.sales,
  sales.quantity,
  sales.price,
  crm.product_cost,
  sales.price - crm.product_cost as profit
FROM baraa_projectwork.silver.crm_sales_details AS sales
LEFT JOIN baraa_projectwork.gold.dim_customers AS customers
  ON customers.customer_id = sales.customer_id
LEFT JOIN baraa_projectwork.gold.dim_products AS products
  ON products.product_number = sales.product_number
LEFT JOIN baraa_projectwork.silver.crm_products as crm
  ON products.product_number = crm.product_number
"""

df = spark.sql(query)
```

```python
df.limit(10).display()
```

```python
df.write.mode("overwrite").format("delta").saveAsTable("baraa_projectwork.gold.fact_sales")
```
