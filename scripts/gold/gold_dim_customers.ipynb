```python
query = """
SELECT
  ROW_NUMBER() OVER(ORDER BY crm.customer_id) as customer_key,
  crm.customer_id,
  erp.customer_number,
  crm.first_name,
  crm.last_name,
  crm.marital_status,
  CASE WHEN crm.gender <> 'n/a' THEN crm.gender
    ELSE coalesce(erp.gender, 'n/a')
  END AS gender,
  erp.birth_date,
  loc.country,
  crm.created_date
FROM baraa_projectwork.silver.crm_customers AS crm
LEFT JOIN baraa_projectwork.silver.erp_customers AS erp
  ON crm.customer_number = erp.customer_number
LEFT JOIN baraa_projectwork.silver.erp_customer_location AS loc
  ON erp.customer_number = loc.customer_number
"""
df = spark.sql(query)
```

```python
df.limit(10).display()
```

```python
df.write.mode("overwrite").format("delta").saveAsTable("baraa_projectwork.gold.dim_customers")
```
